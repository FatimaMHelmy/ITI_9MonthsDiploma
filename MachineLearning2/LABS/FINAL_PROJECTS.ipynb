### DATA NORMALIZATION 
We will try 2 methods and get the best one 

1: MinMaxScaler 

2: RobustScaler 

```python
# 1: MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df)
```

```python
# 2: RobustScaler
from sklearn.preprocessing import RobustScaler
scaler_rob = RobustScaler()
df_rob_scaled = scaler_rob.fit_transform(df)
```

## FEATURE SELECTION USING PCA 
###  FROM THE HRAT MAP IT SUPPOSED TAHT THERE ARE 6 STRONG INDEPENDENT VARIABLES THEN FEATURES WILL BE DECOMPOSED TO 15 


###  PCA WITH SCALED DATA BY MIN MAX 

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=14)
pca.fit(df_scaled)
```

```python
print(pca.explained_variance_)
```

###  PCA WITH SCALED DATA BY ROBUST 

```python
pca = PCA(n_components=14)
pca.fit(df_rob_scaled)
```

```python
print(pca.explained_variance_)
```

### FROM variance  we found that RobustScaler give higher variance and there are 9 components with high variance 
so repeat the PCA with n_components = 9 and with df_rob_scaled

```python
pca_9comp = PCA(n_components=0.95)
pca_9comp.fit(df_rob_scaled)
print(pca_9comp.components_)
```

```python
print(pca_9comp.explained_variance_)
```

## **from the previous steps the new data is pca_9comp.components_**

```python
new_df = pca_9comp.transform(df_rob_scaled)
```

```python
new_df.shape
```

##  let's apply CLUSTERING technique 

### **1: K_MEANS **

```python
## make instance of the model 
X = new_df
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)
labels = kmeans.labels_
print("Score ==",silhouette_score(X, labels))
```

## *DBSCAN*

```python
X = new_df.copy()
db = DBSCAN(eps = 3.4,min_samples = 5).fit(X)
labels = db.labels_
print("score of clustered data == ",silhouette_score(df, labels))
```

```python
set(list(labels))
```

### Hierarchical

```python
## make instance of the model 
model = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
X = new_df.copy()
d_hir = model.fit(X)
labels = model.labels_
print("SCORE == ",silhouette_score(X, labels))
```

## GaussianMixture

```python
X = new_df.copy()
gm = GaussianMixture(n_components=2, random_state=0).fit(X)
labels = gm.predict(X)
print("Score == ",silhouette_score(X, labels))
```
